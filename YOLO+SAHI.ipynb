```python
from IPython.display import Image

from sahi import AutoDetectionModel
from sahi.predict import get_prediction, get_sliced_prediction, predict
from sahi.utils.cv import read_image
```

```python
detection_model = AutoDetectionModel.from_pretrained(
    model_type="ultralytics",
    model_path="yolo11n.pt",
    confidence_threshold=0.35,
    device="cpu",
)
```

```python
result = get_prediction(read_image("/content/drive/MyDrive/Dataset/small-vehicles1.jpeg"), detection_model)
```

```python
result.export_visuals(export_dir="/content/drive/MyDrive/Dataset/", hide_conf=True)

Image("/content/drive/MyDrive/Dataset/prediction_visual.png")
```

```python
result = get_sliced_prediction(
    "/content/drive/MyDrive/Dataset/small-vehicles1.jpeg",
    detection_model,
    slice_height=256,
    slice_width=256,
    overlap_height_ratio=0.2,
    overlap_width_ratio=0.2,
)
```

```python
result.export_visuals(export_dir="/content/drive/MyDrive/Dataset/", hide_conf=True)

Image("/content/drive/MyDrive/Dataset/prediction_visual.png")
```

```python
object_prediction_list = result.object_prediction_list
print(object_prediction_list)
```
